Defining HITS function that accepts an Adjacency matrix as argument and calculates the authority and hub scores of websites.(Assuming 5 iterations)

```python
def hits(A):
    n = len(A)
    Au = dot(transpose(A),A)
    Hu = dot(A,transpose(A))
    a = ones(n); h = ones(n)
    for j in range(5):
        a = dot(a,Au)
        a = a/sum(a)
        h = dot(h,Hu)
        h = h/sum(h)
    
    label = websites[0:n] 
    
    index = np.arange(len(label))
    plt.bar(index, a)
    plt.xlabel('Websites', fontsize=10)
    plt.ylabel('Authority Score', fontsize=10)
    plt.xticks(index, label, fontsize=7, rotation=90)
    plt.title('Authority')
    plt.show()
    
    index = np.arange(len(label))
    plt.bar(index, h)
    plt.xlabel('Websites', fontsize=10)
    plt.ylabel('Hub Score', fontsize=10)
    plt.xticks(index, label, fontsize=7, rotation=90)
    plt.title('Hub')
    plt.show()
```
